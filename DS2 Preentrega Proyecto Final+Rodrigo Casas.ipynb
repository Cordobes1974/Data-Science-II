```python
import requests
import pandas as pd
from tqdm import tqdm
import time  # For adding delay

def obtener_datos_api(url, headers=None, nombre_fuente="API"):
    """Realiza una solicitud a una API y devuelve los datos como un DataFrame."""
    try:
        print(f"Obteniendo datos de {nombre_fuente}: {url}...") # Debugging statement # Changed to print both values
        response = requests.get(url, headers=headers, timeout=10)  # Agregar timeout
        print(f"Response status code: {response.status_code}")  # Debugging statement
        response.raise_for_status()  # Elevar excepción si hay error HTTP
        data = response.json()
        print("Data fetched successfully")  # Added for debugging

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return pd.DataFrame()

    # Uso tqdm para mostrar el progreso mientras se está creando el DataFrame
    df = pd.DataFrame(tqdm(data, desc=f"Creating DataFrame from {nombre_fuente}"))
    print(f"Datos obtenidos de {nombre_fuente} correctamente.")
    return df
```

# **Uniendo y cruzando DataFrames con merge() y concat()**
Ahora que tenemos los DataFrames del BCRA, almacenados en el diccionario bcra_data, podemos usar merge() y concat(), para combinarlos y obtener información más valiosa.

merge()
La función merge() se usa para combinar DataFrames basándose en una o más columnas comunes. Es similar a una operación "JOIN" en SQL.

Ejemplo:

```python
# Supongamos que queremos unir 'usd_df' y 'usd_of_df' por la columna 'd' (fecha)
merged_df = pd.merge(bcra_data['usd'], bcra_data['usd_of'], on='d', how='inner')

# 'on='d'' especifica la columna común para unir
# 'how='inner'' especifica que solo se incluirán las filas con valores coincidentes en 'd'
```

Opciones de how:

* inner: Solo filas con valores coincidentes en ambas DataFrames.

* outer: Todas las filas de ambos DataFrames, rellenando con NaN donde no haya coincidencias.

* left: Todas las filas del DataFrame izquierdo y las filas coincidentes del DataFrame derecho.

* right: Todas las filas del DataFrame derecho y las filas coincidentes del DataFrame izquierdo.

concat()

La función concat() se usa para concatenar o apilar DataFrames. Puedes concatenar verticalmente (añadiendo filas) u horizontalmente (añadiendo columnas).

Ejemplo:

```python
# Concatenar verticalmente 'base_df' y 'reservas_df'
concatenated_df = pd.concat([bcra_data['base'], bcra_data['reservas']], axis=0)

# 'axis=0' especifica concatenación vertical (por filas)
```

Opciones de axis:

0: Concatenación vertical (por filas).
1: Concatenación horizontal (por columnas).

```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

class ArgentinaEconomicData:
    def __init__(self):
        self.bcra_base_url = "http://api.estadisticas.bcra.gob.ar/api"
        self.bcra_token = "eyJhbGciOiJIUzUxMiIsInR5cCI6IkpXVCJ9.eyJleHAiOjE3NjY5NjYyMzMsInR5cGUiOiJleHRlcm5hbCIsInVzZXIiOiJyY2FzYXNzaWdsb3h4aUBnbWFpbC5jb20ifQ.UVvUHmySmkIDdL_1HtSoCcRweJucCKjNUDBkY833ovgu9xbzSzkOEoYGvdU7OcnhJfrHa6Ke3Q7uQN_l-TnR8Q"
        self.datos_argentina_url = "https://apis.datos.gob.ar/series/api"

    def get_bcra_data(self, indicator_id, start_date=None, end_date=None):
        """
        Obtiene datos del BCRA para un indicador específico
        """
        headers = {'Authorization': f'Bearer {self.bcra_token}'}

        if not start_date:
            start_date = (datetime.now() - timedelta(days=365)).strftime('%Y-%m-%d')
        if not end_date:
            end_date = datetime.now().strftime('%Y-%m-%d')

        endpoint = f"{self.bcra_base_url}/series/{indicator_id}/datos/{start_date}/{end_date}"

        try:
            response = requests.get(endpoint, headers=headers)
            response.raise_for_status()
            return pd.DataFrame(response.json())
        except requests.exceptions.RequestException as e:
            print(f"Error al obtener datos del BCRA: {e}")
            return None

    def get_economic_activity_data(self):
        """
        Obtiene datos de actividad económica por sector
        """
        series_ids = {
            'agricultura': 'agricultura_id',
            'industria': 'industria_id',
            'comercio': 'comercio_id',
            'construccion': 'construccion_id',
            'servicios_financieros': 'servicios_id'
        }

        params = {
            'ids': ','.join(series_ids.values()),
            'format': 'json',
            'limit': 1000
        }

        try:
            response = requests.get(f"{self.datos_argentina_url}/series", params=params)
            response.raise_for_status()
            return pd.DataFrame(response.json()['data'])
        except requests.exceptions.RequestException as e:
            print(f"Error al obtener datos de actividad económica: {e}")
            return None

class FinancialSegmentationAnalyzer:
    def __init__(self, economic_data_handler):
        self.data_handler = economic_data_handler
        self.segments = {}
        self.scaler = StandardScaler()

    def prepare_company_data(self, company_data):
        """
        Prepara los datos de las empresas para la segmentación
        """
        self.size_criteria = {
            'micro': {'max_empleados': 15, 'max_ingresos': 50000000},
            'pequeña': {'max_empleados': 60, 'max_ingresos': 400000000},
            'mediana': {'max_empleados': 235, 'max_ingresos': 2000000000}
        }

        return company_data

    def segment_by_size(self, company_data):
        """Segmenta empresas por tamaño según criterios PyME"""
        def determine_size(row):
            if row['ingresos_anuales'] <= self.size_criteria['micro']['max_ingresos']:
                return 'micro'
            elif row['ingresos_anuales'] <= self.size_criteria['pequeña']['max_ingresos']:
                return 'pequeña'
            elif row['ingresos_anuales'] <= self.size_criteria['mediana']['max_ingresos']:
                return 'mediana'
            return 'grande'

        company_data['segmento_tamaño'] = company_data.apply(determine_size, axis=1)
        return company_data

    def analyze_financial_capacity(self, company_data):
        """Analiza la capacidad financiera basada en múltiples factores"""
        financial_vars = ['ingresos_anuales', 'deuda_actual']
        scaled_data = self.scaler.fit_transform(company_data[financial_vars])

        company_data['capacidad_financiera'] = np.mean(scaled_data, axis=1)
        company_data['nivel_capacidad'] = pd.qcut(
            company_data['capacidad_financiera'],
            q=3,
            labels=['baja', 'media', 'alta']
        )

        return company_data

    def recommend_products(self, segment_data):
        """Recomienda productos financieros basados en la segmentación"""
        productos_recomendados = {
            'micro': {
                'baja': [
                    'Microcréditos productivos',
                    'Descuento de cheques',
                    'Tarjeta de crédito empresarial básica'
                ],
                'media': [
                    'Préstamos para capital de trabajo',
                    'Leasing de equipamiento',
                    'Cuenta corriente con acuerdo'
                ],
                'alta': [
                    'Préstamos para inversión',
                    'Factoring',
                    'Paquete financiero integral'
                ]
            },
            'pequeña': {
                'baja': [
                    'Préstamos con garantía SGR',
                    'Descuento de cheques',
                    'Préstamos para capital de trabajo'
                ],
                'media': [
                    'Leasing',
                    'Préstamos para inversión productiva',
                    'Comercio exterior básico'
                ],
                'alta': [
                    'Financiación estructurada',
                    'Comercio exterior ampliado',
                    'Cobertura de riesgo cambiario'
                ]
            },
            'mediana': {
                'baja': [
                    'Préstamos con garantía SGR',
                    'Financiación de capital de trabajo',
                    'Leasing'
                ],
                'media': [
                    'Financiación de proyectos',
                    'Comercio exterior',
                    'Inversiones estructuradas'
                ],
                'alta': [
                    'Mercado de capitales',
                    'Servicios de tesorería',
                    'Financiación sindicada'
                ]
            }
        }

        recommendations = []
        for _, company in segment_data.iterrows():
            size = company['segmento_tamaño']
            capacity = company['nivel_capacidad']
            sector = company['sector']

            if size in productos_recomendados and capacity in productos_recomendados[size]:
                products = productos_recomendados[size][capacity]
                recommendations.append({
                    'sector': sector,
                    'tamaño': size,
                    'capacidad': capacity,
                    'productos_recomendados': products
                })

        return pd.DataFrame(recommendations)

    def generate_market_insights(self, segment_data):
        """Genera insights del mercado basados en la segmentación"""
        insights = {
            'distribución_segmentos': segment_data['segmento_tamaño'].value_counts(),
            'capacidad_por_sector': pd.crosstab(
                segment_data['sector'],
                segment_data['nivel_capacidad']
            ),
            'oportunidades_sector': segment_data.groupby('sector')['capacidad_financiera'].mean()
        }

        return insights

# Ejemplo de uso
def main():
    # Crear instancias
    economic_data = ArgentinaEconomicData()
    analyzer = FinancialSegmentationAnalyzer(economic_data)

    # Datos de ejemplo (reemplazar con datos reales)
    sample_data = pd.DataFrame({
        'sector': ['Agropecuario', 'Industrial', 'Comercio'],
        'ingresos_anuales': [30000000, 450000000, 1500000000],
        'empleados': [10, 50, 150],
        'antiguedad': [5, 15, 25],
        'deuda_actual': [5000000, 100000000, 500000000]
    })

    # Procesar datos
    processed_data = analyzer.prepare_company_data(sample_data)
    segmented_data = analyzer.segment_by_size(processed_data)
    analyzed_data = analyzer.analyze_financial_capacity(segmented_data)

    # Obtener recomendaciones
    recommendations = analyzer.recommend_products(analyzed_data)
    print("\nRecomendaciones de productos:")
    print(recommendations)

    # Generar insights
    market_insights = analyzer.generate_market_insights(analyzed_data)
    print("\nInsights del mercado:")
    for key, value in market_insights.items():
        print(f"\n{key}:")
        print(value)

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from sklearn.preprocessing import StandardScaler

class ArgentinaEconomicData:
    def __init__(self):
        self.bcra_base_url = "http://api.estadisticas.bcra.gob.ar/api"
        self.bcra_token = "eyJhbGciOiJIUzUxMiIsInR5cCI6IkpXVCJ9.eyJleHAiOjE3NjY5NjYyMzMsInR5cGUiOiJleHRlcm5hbCIsInVzZXIiOiJyY2FzYXNzaWdsb3h4aUBnbWFpbC5jb20ifQ.UVvUHmySmkIDdL_1HtSoCcRweJucCKjNUDBkY833ovgu9xbzSzkOEoYGvdU7OcnhJfrHa6Ke3Q7uQN_l-TnR8Q"
        self.datos_argentina_url = "https://apis.datos.gob.ar/series/api"

    def get_economic_activity_data(self):
        """
        Obtiene datos de actividad económica usando IDs válidos de datos.gob.ar
        """
        # IDs válidos de series relacionadas con actividad económica
        # Estos son ejemplos - deberás reemplazarlos con IDs reales de tu interés
        series_ids = {
            'emae': '143.3_NO_PR_2004_A_21',  # EMAE (Estimador Mensual de Actividad Económica)
            'ipi': '11.3_IPIA_0_M_20',        # IPI (Índice de Producción Industrial)
            'isac': '11.3_ISAC_2012_M_23',    # ISAC (Indicador Sintético de la Actividad de la Construcción)
        }

        params = {
            'ids': ','.join(series_ids.values()),
            'format': 'json',
            'limit': 1000
        }

        try:
            response = requests.get(f"{self.datos_argentina_url}/series", params=params)
            response.raise_for_status()
            data = response.json()

            if 'data' in data:
                return pd.DataFrame(data['data'])
            else:
                print("No se encontraron datos en la respuesta")
                return pd.DataFrame()

        except requests.exceptions.RequestException as e:
            print(f"Error al obtener datos de actividad económica: {e}")
            return pd.DataFrame()

    def get_bcra_data(self, indicator_id, start_date=None, end_date=None):
        """
        Obtiene datos del BCRA con mejor manejo de errores
        """
        if not self.bcra_token or self.bcra_token == "YOUR_TOKEN_HERE":
            print("Token del BCRA no configurado")
            return pd.DataFrame()

        headers = {'Authorization': f'Bearer {self.bcra_token}'}

        if not start_date:
            start_date = (datetime.now() - timedelta(days=365)).strftime('%Y-%m-%d')
        if not end_date:
            end_date = datetime.now().strftime('%Y-%m-%d')

        try:
            response = requests.get(
                f"{self.bcra_base_url}/series/{indicator_id}/datos/{start_date}/{end_date}",
                headers=headers
            )
            response.raise_for_status()
            return pd.DataFrame(response.json())
        except requests.exceptions.RequestException as e:
            print(f"Error al obtener datos del BCRA: {e}")
            return pd.DataFrame()

class IERALDataCollector:
    def __init__(self):
        # Cambiamos a la URL alternativa
        self.base_url = "https://www.fundmediterranea.org.ar"

    def get_latest_reports(self, limit=5):
        """
        Obtiene informes con mejor manejo de errores
        """
        try:
            response = requests.get(self.base_url, timeout=10)
            response.raise_for_status()

            # Crear DataFrame vacío con la estructura esperada
            empty_df = pd.DataFrame({
                'titulo': [],
                'fecha': [],
                'url': [],
                'resumen': []
            })

            soup = BeautifulSoup(response.content, 'html.parser')
            reports = []

            # Buscar los artículos de informes
            articles = soup.find_all('article', limit=limit)

            for article in articles:
                report = {
                    'titulo': article.find('h2').text.strip() if article.find('h2') else 'Sin título',
                    'fecha': self._parse_date(article.find('time').text.strip() if article.find('time') else ''),
                    'url': article.find('a')['href'] if article.find('a') else None,
                    'resumen': article.find('p').text.strip() if article.find('p') else 'Sin resumen'
                }
                reports.append(report)

            if reports:
                return pd.DataFrame(reports)
            return empty_df

        except requests.exceptions.RequestException as e:
            print(f"Error al obtener informes de IERAL: {e}")
            return pd.DataFrame({
                'titulo': [],
                'fecha': [],
                'url': [],
                'resumen': []
            })

    def _parse_date(self, date_str):
        """Parsea fechas con mejor manejo de errores"""
        try:
            return pd.to_datetime(date_str, format='%d/%m/%Y')
        except:
            return pd.NaT

class MarketAnalyzer:
    def __init__(self):
        self.economic_data = ArgentinaEconomicData()
        self.ieral_collector = IERALDataCollector()

    def analyze_market(self, sector=None):
        """
        Realiza un análisis de mercado combinando múltiples fuentes
        """
        # Obtener datos económicos
        economic_data = self.economic_data.get_economic_activity_data()

        # Obtener informes de IERAL
        ieral_reports = self.ieral_collector.get_latest_reports()

        # Filtrar por sector si se especifica
        if sector and not ieral_reports.empty:
            sector_reports = ieral_reports[
                ieral_reports['titulo'].str.contains(str(sector), case=False, na=False) |
                ieral_reports['resumen'].str.contains(str(sector), case=False, na=False)
            ]
        else:
            sector_reports = ieral_reports

        return {
            'datos_economicos': economic_data,
            'informes_sectoriales': sector_reports,
            'fecha_analisis': datetime.now().strftime('%Y-%m-%d')
        }

    def generate_report(self, analysis_data):
        """
        Genera un reporte basado en los datos analizados
        """
        report = {
            'fecha': analysis_data['fecha_analisis'],
            'indicadores_disponibles': len(analysis_data['datos_economicos'].columns) if not analysis_data['datos_economicos'].empty else 0,
            'informes_analizados': len(analysis_data['informes_sectoriales']),
        }

        return report

def main():
    try:
        # Crear analizador
        analyzer = MarketAnalyzer()

        # Realizar análisis
        analysis = analyzer.analyze_market(sector="Agropecuario")

        # Generar reporte
        report = analyzer.generate_report(analysis)

        # Mostrar resultados
        print("\nReporte de Análisis de Mercado:")
        print(f"Fecha: {report['fecha']}")
        print(f"Indicadores económicos disponibles: {report['indicadores_disponibles']}")
        print(f"Informes sectoriales analizados: {report['informes_analizados']}")

        if not analysis['datos_economicos'].empty:
            print("\nIndicadores económicos disponibles:")
            print(analysis['datos_economicos'].columns.tolist())

        if not analysis['informes_sectoriales'].empty:
            print("\nÚltimos informes disponibles:")
            for _, informe in analysis['informes_sectoriales'].iterrows():
                print(f"- {informe['titulo']}")

    except Exception as e:
        print(f"\nError en el análisis: {str(e)}")
        print("El programa continuará con los datos disponibles")

if __name__ == "__main__":
    main()
```

**Código para web scraping y procesamiento de texto, para extraer información de la página web de IERAL**

```python
!pip install requests beautifulsoup4 PyPDF2
import requests
from bs4 import BeautifulSoup
import PyPDF2
import re
import os

def scrape_ieral_reports(url, num_pages=1):
    """
    Realiza web scraping de la página de informes de IERAL y extrae información.

    Args:
        url (str): URL de la página de informes de IERAL.
        num_pages (int): Número de páginas a analizar (por defecto, 1).

    Returns:
        list: Una lista de diccionarios, donde cada diccionario representa un informe.
    """
    reports = []
    for page_num in range(1, num_pages + 1):
        page_url = f"{url}/novedades.asp?seccion=40&pagina={page_num}"  # URL con paginación
        try:
            response = requests.get(page_url)
            response.raise_for_status()
            soup = BeautifulSoup(response.content, 'html.parser')

            # Encuentra los elementos que contienen la información de los informes
            report_elements = soup.find_all('div', class_='novedad')

            for element in report_elements:
                report_data = {}
                report_data['titulo'] = element.find('h3').text.strip()
                report_data['fecha'] = element.find('span', class_='fecha').text.strip()
                report_data['resumen'] = element.find('p').text.strip()
                report_data['url'] = element.find('a')['href']

                # Descarga y procesa el PDF si está disponible
                if report_data['url'].endswith('.pdf'):
                    pdf_url = f"{url}/{report_data['url']}"
                    pdf_content = download_pdf(pdf_url)
                    if pdf_content:
                        report_data['texto_completo'] = extract_text_from_pdf(pdf_content)

                reports.append(report_data)

        except requests.exceptions.RequestException as e:
            print(f"Error al obtener datos de la página {page_num}: {e}")
        except Exception as e:
            print(f"Error al procesar un informe en la página {page_num}: {e}")

    return reports


def download_pdf(url):
    """Descarga un PDF desde una URL."""
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()
        return response.content
    except requests.exceptions.RequestException as e:
        print(f"Error al descargar PDF: {e}")
        return None


def extract_text_from_pdf(pdf_content):
    """Extrae texto de un contenido PDF."""
    try:
        pdf_reader = PyPDF2.PdfReader(pdf_content)
        text = ""
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
        return text
    except Exception as e:
        print(f"Error al extraer texto del PDF: {e}")
        return None


def preprocess_text(text):
    """Realiza preprocesamiento del texto."""
    # 1. Convertir a minúsculas
    text = text.lower()
    # 2. Eliminar caracteres especiales y números
    text = re.sub(r"[^a-záéíóúüñ\s]", "", text)
    # 3. Eliminar espacios en blanco adicionales
    text = re.sub(r"\s+", " ", text)
    return text


def main():
    ieral_url = "https://www.ieral.org"
    reports = scrape_ieral_reports(ieral_url, num_pages=2)  # Analiza 2 páginas
    # Aquí puedes procesar 'reports', que contiene la información extraída
    print(reports)

if __name__ == "__main__":
    main()
```

Explicación del Código:

1. scrape_ieral_reports:

*Recibe la URL de la página de informes de IERAL y el número de páginas a analizar.
* Realiza web scraping usando requests y Beautiful Soup.
* Extrae información como título, fecha, resumen y URL del informe.
* Si el informe es un PDF, lo descarga y extrae el texto.
* Guarda la información en una lista de diccionarios.
2. download_pdf: Descarga un PDF desde la URL dada.

3. extract_text_from_pdf: Extrae texto de un contenido PDF usando PyPDF2.

4. preprocess_text: Realiza un preprocesamiento básico del texto: convierte a minúsculas, elimina caracteres especiales y números, y elimina espacios en blanco adicionales.

5. main:

* Define la URL base de IERAL.
* Llama a scrape_ieral_reports para obtener la información de los informes.
* Imprime la información extraída.
